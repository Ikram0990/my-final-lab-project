```python

import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10,10,200)
y = x ** 2
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y= x**2")
plt.title("График параболы y= x**2")
plt.grid(True)
plt.show()
```
